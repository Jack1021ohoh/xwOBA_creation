```python
from pybaseball import statcast, cache, statcast_batter_expected_stats
```

```python
data = statcast('2024-03-28', '2024-09-30')
```

```python
data['events'].unique()
```

```python
data = data.dropna(subset = ['events'])
print(data.shape)
```

```python
data.to_csv('2024_hit_data.csv', index = False)
```

```python
data2 = statcast('2023-03-30', '2023-10-02')
```
